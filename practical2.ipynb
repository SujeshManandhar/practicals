```python
#training and testing data->80% data for training, 10% for validation and 10% for test
training_data, holdout_data=sklearn.model_selection.train_test_split(salaries,test_size=0.2)
validation_data, test_data=sklearn.model_selection.train_test_split(holdout_data,test_size=0.5)
```

```python
training_data.shape,validation_data.shape,test_data.shape
```

```python
#regression using sckitlearn
#creating regresson object

#training the model in training data. needs x and y value
xtrain=training_data[['AnnualSalary']] #feature for training
ytrain=training_data['GrossPay'] #target for training
xval=validation_data[['AnnualSalary']]  #feature for valdation
yval=validation_data['GrossPay'] #what to predict #target for validation
reg=sklearn.linear_model.LinearRegression() #creating linear regression model
reg.fit(xtrain,ytrain) #train the model in training data
#prdicting in validation data. it uses trained model to predict
val_predic = reg.predict(xval)
#displaying
print("slope:",reg.coef_[0])
print("intercept:",reg.intercept_)
```

```python
#theil sen regression
ts = sklearn.linear_model.TheilSenRegressor()
ts.fit(xtrain,ytrain)
tsPredict = ts.predict(xval)
print("slope:",ts.coef_[0])
```

```python
#displaying regression line and actual value 
plt.scatter(xval,yval,color='blue', label='actual validation data')
plt.plot(xval,tsPredict,color='red', linewidth=2,label='regression')
plt.xlabel('annual salary')
plt.ylabel('gross salary')
plt.legend()
```

```python
#Ransac regressor
ransac=sklearn.linear_model.RANSACRegressor()
ransac.fit(xtrain,ytrain)
ransacPredict=ransac.predict(xval)
print("slope:",ransac.estimator_.coef_)
```

```python
#displaying in diagram
plt.scatter(xval,yval,color='blue', label='actual validation data')
plt.plot(xval,ransacPredict,color='red', linewidth=2,label='regression')
plt.xlabel('annual salary')
plt.ylabel('gross salary')
plt.legend()
```

```python
#huber model
huber = sklearn.linear_model.HuberRegressor()
huber.fit(xtrain,ytrain)
huberpredict=huber.predict(xval)
print("slope:",huber.coef_[0])
```

```python
#   "Now imagine that you are the City of Baltimore's hiring manager. New employees regularly ask how much they are actually likely to earn given the salary to which they are about to agree.\n",
 #   "You don't want to give an answer that is too high because you might be putting the city at risk for a lawsuit for misrepresenting the job. On the other hand, you don't want to give an answer that's too low because the candidate might pass up on the job and work elsewhere.\n",
  #  "\n",
   # "You decide that it will cost \\\\$0.05 in lawsuit risk for each dollar you overrepresent, but only \\\\$0.01 for each dollar you underrepresent.\n",
    #"\n",
    #"E.g., if a candidate is actually likely to earn \\\\$100,000 and you say \\\\$120,000, this is worth \\\\$10,000 in potential lawsuits for misrepresentation. If you say \\\\$80,000, then that will cost you \\$200 in potential recruiters' fees to find someone else.\n",
    #"\n",
    #"Here's a function that will calculate the cost of these predictions. Don't worry if it is too complicated for you to create\n",
    #"by yourself right now, but see if you can make sense of how it works"
```

```python
#defiining function
def cost(actuals,predictions):
    cost=0.0
    for actual,prediction in zip(actuals,predictions):
        if prediction>actual:
            cost+=0.05*(prediction-actual)
        elif prediction<actual:
            cost+=0.01*(actual-prediction)
        else:
            cost+=0.0
    return cost
```
